# signal denoising
just mean of repeating noised signals

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [5, 5]
```

```python
n = 1024
seconds = 10
t = np.linspace(0, seconds, 1024)
signal = np.sin(t) # clean signal 
plt.plot(t, signal)
```

```python
noise_radius = 1.1 # 1 means random -1..1, -noise_radius..noise_radius
noise_signals_n = 2000
noise = 2 * noise_radius * np.random.random((noise_signals_n, n)) - noise_radius
plt.plot(t, noise[20])
# plt.imshow(noise)
```

```python
noised_signals = signal + noise
plt.plot(t, noise[20], 'g')
plt.plot(t, noised_signals[20], 'y', linewidth=0.7)
plt.plot(t, signal, 'r')
```

```python
plt.plot(t, np.mean(noised_signals, axis=0))
```
